```python
# Soal 1 - Binary Search

def search(nums, target):
    # Inisialisasi pointer untuk batas bawah (low) dan batas atas (high)
    low = 0
    high = len(nums) - 1

    # Lakukan perulangan selama batas bawah tidak melewati batas atas
    while low <= high:
        # Hitung indeks tengah. Menggunakan (low + high) // 2 untuk integer division.
        # Atau low + (high - low) // 2 untuk menghindari potensi overflow (meskipun jarang di Python)
        mid = low + (high - low) // 2

        # Cek apakah elemen di tengah adalah target
        if nums[mid] == target:
            # Jika ditemukan, kembalikan indeksnya
            return mid
        # Jika elemen di tengah lebih kecil dari target,
        # cari di bagian kanan (abaikan mid dan kiri)
        elif nums[mid] < target:
            low = mid + 1
        # Jika elemen di tengah lebih besar dari target,
        # cari di bagian kiri (abaikan mid dan kanan)
        else:
            high = mid - 1

    # Jika perulangan selesai dan target tidak ditemukan
    return -1

# Contoh Input dan Output
nums1 = [-1, 0, 3, 5, 9, 12]
target1 = 9
result1 = search(nums1, target1)
print(f"Input: nums = {nums1}, target = {target1}")
print(f"Output: {result1}") # Output: 4 [cite: 7]

nums2 = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target2 = 56
result2 = search(nums2, target2)
print(f"Input: nums = {nums2}, target = {target2}")
print(f"Output: {result2}") # Output: 7

nums3 = [2, 5, 8, 12]
target3 = 10
result3 = search(nums3, target3)
print(f"Input: nums = {nums3}, target = {target3}")
print(f"Output: {result3}") # Output: -1
```

```python
# Soal 2 - Squares of a Sorted Array

def sortedSquares(nums):
    n = len(nums)
    # Buat array hasil dengan ukuran n, diinisialisasi dengan nol
    result = [0] * n

    # Inisialisasi pointer kiri (left), kanan (right), dan pointer untuk mengisi array hasil (k)
    left = 0
    right = n - 1
    # Pointer k dimulai dari indeks terakhir (n-1) karena kita akan mengisi dari nilai terbesar
    k = n - 1

    # Lakukan perulangan selama pointer kiri tidak melewati pointer kanan
    while left <= right:
        # Bandingkan nilai mutlak dari elemen di kiri dan kanan
        # Kuadrat dari nilai mutlak terbesar akan menjadi nilai terbesar berikutnya
        if abs(nums[left]) >= abs(nums[right]):
            # Kuadrat dari elemen kiri lebih besar (atau sama)
            square = nums[left] * nums[left]
            # Geser pointer kiri ke kanan
            left += 1
        else:
            # Kuadrat dari elemen kanan lebih besar
            square = nums[right] * nums[right]
            # Geser pointer kanan ke kiri
            right -= 1

        # Simpan kuadrat terbesar di posisi k, lalu geser k ke kiri
        result[k] = square
        k -= 1

    # Kembalikan array hasil yang sudah terurut naik
    return result

# Contoh Input dan Output
nums1 = [-4, -1, 0, 3, 10]
result1 = sortedSquares(nums1)
print(f"Input: nums = {nums1}")
print(f"Output: {result1}") # Output: [0, 1, 9, 16, 100] [cite: 13]

nums2 = [-7, -3, 2, 3, 11]
result2 = sortedSquares(nums2)
print(f"Input: nums = {nums2}")
print(f"Output: {result2}") # Output: [4, 9, 9, 49, 121]
```

```python
# Soal 3 - Merge Two Sorted Lists

def mergeTwoSortedLists(list1, list2):
    # Inisialisasi list baru untuk menampung hasil gabungan
    merged_list = []

    # Inisialisasi pointer untuk list1 (i) dan list2 (j)
    i = 0
    j = 0

    # Panjang kedua list
    n1 = len(list1)
    n2 = len(list2)

    # Lakukan perulangan selama kedua list masih memiliki elemen
    while i < n1 and j < n2:
        # Bandingkan elemen saat ini dari kedua list
        if list1[i] <= list2[j]:
            # Jika elemen list1 lebih kecil (atau sama), tambahkan ke merged_list
            merged_list.append(list1[i])
            # Pindah ke elemen berikutnya di list1
            i += 1
        else:
            # Jika elemen list2 lebih kecil, tambahkan ke merged_list
            merged_list.append(list2[j])
            # Pindah ke elemen berikutnya di list2
            j += 1

    # Setelah salah satu list habis, tambahkan sisa elemen dari list1
    # (Perulangan ini akan berjalan hanya jika i < n1)
    while i < n1:
        merged_list.append(list1[i])
        i += 1

    # Tambahkan sisa elemen dari list2
    # (Perulangan ini akan berjalan hanya jika j < n2)
    while j < n2:
        merged_list.append(list2[j])
        j += 1

    # Kembalikan list yang sudah digabungkan dan terurut
    return merged_list

# Contoh Input dan Output
list_a = [1, 2, 4]
list_b = [1, 3, 4]
result_ab = mergeTwoSortedLists(list_a, list_b)
print(f"Input: list1 = {list_a}, list2 = {list_b}")
print(f"Output: {result_ab}") # Output: [1, 1, 2, 3, 4, 4] [cite: 17]

list_c = [5, 6, 7]
list_d = [1, 2, 3]
result_cd = mergeTwoSortedLists(list_c, list_d)
print(f"Input: list1 = {list_c}, list2 = {list_d}")
print(f"Output: {result_cd}") # Output: [1, 2, 3, 5, 6, 7]
```
